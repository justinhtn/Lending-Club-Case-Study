### Data exploration

##### Reviewing columns and row data 

```python
# import data

data = pd.read_csv('loan.csv')
```

```python
data.info()
```

```python
data.shape
```

```python
data.head()
```

```python
data.describe()
```

```python
# reviewing disribution of null values in columns

col_null_values = data.isnull().sum()
col_null_values.hist(nbins=20)
```

```python
# reviewing columns with high null values

null_val_limit = 12000
col_null_values[col_null_values.values >= null_val_limit ]
```

```python
# creating a list of columns with high null values

cols_to_remove = col_null_values[col_null_values.values >= null_val_limit ].keys()
```

```python
# dropping columns with high null values

data.drop(columns= cols_to_remove, inplace=True)
```

```python
# checking to ensure columns were dropped.

data.isnull().sum().sort_values(ascending=False)
```

```python
# checking to ensure distribution of null values is low.

data.isnull().sum().hist()
```

```python
# reviewing number of unique items in columns. 

data.nunique().sort_values(ascending=True)
```

```python
# pulling column names for columns that have only 1 unique value

single_val_cols = data.nunique()[data.nunique().values == 1].keys()
single_val_cols
```

```python
# dropping single value columns

data.drop(columns=single_val_cols, inplace=True)
```

```python
# checking data shape to ensure columns were dropped

data.shape
```

#### Let's look to see the description of columns we have left

```python
# pull in data dictionary for column evaluation and transposing so we can filter by column

data_dict = pd.read_excel('Data_Dictionary.xlsx', index_col=0, header=None).iloc[:, 0:1].transpose()
data_dict
```

```python
# filter by just column we have kept and re-tranposing for easy visibility
data_dict = data_dict[list(data.columns)].transpose()
data_dict.head()
```

##### Feature transformation

```python
data.head()
```

```python
def get_quick_facts(column):

    print(f"-----Description of {column}-----\n",data_dict.loc[column][1],"\n")

    print("-----Head of data-----")
    print(data[column].head(), "\n")

    print("-----Quick facts-----")
    print(data[column].describe(), "\n")

    print("-----Unique-----")
    print(data[column].unique(), "\n")

    print("-----Distribution-----")
    return data[column].hist()
```

```python
get_quick_facts('loan_amnt')
```

```python
get_quick_facts('funded_amnt')
```

```python
get_quick_facts('term')
```

```python
get_quick_facts('int_rate')
```

```python
# removing % from int_rate values
data['int_rate'] = pd.to_numeric(data['int_rate'].str[:-1], errors='coerce')
```

```python
get_quick_facts('int_rate')
```
